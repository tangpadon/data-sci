```python
#Import dataset & library
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_openml
diabetes = fetch_openml(name='diabetes', version=1, as_frame=True)
diabetes.data.shape, diabetes.target.shape
df = diabetes.frame
#check missing value
print(df.isnull().values.any())
```

```python
df.head()
```

```python
#test 30%
Xtrain, Xtest, ytrain, ytest = train_test_split(diabetes.data, diabetes.target, test_size=0.3, random_state=0)
```

```python
#Import Naive Bayes
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
model.predict([[1,148,66,72,35,0,26.6,0.627]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```
