```python
import pandas as pd
data = pd.read_csv('data/Mall_Customers.csv')
df = pd.DataFrame(data)
print(df.isnull().sum())
print(df.head())
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import silhouette_score
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data = df, x = 'Age', y = 'Annual_Income_(k$)', hue = 'Spending_Score')
plt.xlim(15,75)
plt.show()
```

```python
Xtrain= df[['Age','Annual_Income_(k$)']]
```

```python
k = 3
model = KMeans(n_clusters=k).fit(Xtrain)
```

```python
sns.scatterplot(data = Xtrain, x = 'Age', y = 'Annual_Income_(k$)', hue = model.labels_)
plt.show()
```

```python
silhouette_score(Xtrain, model.labels_, metric = 'euclidean')
```
